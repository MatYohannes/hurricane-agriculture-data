## Grouping and Aggregating Data

````python
#Resample the data to calculate year averages
yearly_avg_commodities = filtered_commodities_df.resample('YE').mean()
yearly_avg_commodities['year'] = yearly_avg_commodities.index.year
yearly_avg_commodities
````

````python
commodities_inflation_adjusted = pd.merge(yearly_avg_commodities, filtered_inflation_df[['year', 'inflation_rate']], on='year', how='left')

commodities_inflation_adjusted.set_index('year', inplace=True)
commodities_inflation_adjusted
````

````python
# Adjust each commodity value for inflation by dividing by (1 + inflation_rate/100)
commodities_inflation_adjusted['coffee_value_adjusted'] = commodities_inflation_adjusted['coffee_value'] / (1 + commodities_inflation_adjusted['inflation_rate'] / 100)
commodities_inflation_adjusted['corn_value_adjusted'] = commodities_inflation_adjusted['corn_value'] / (1 + commodities_inflation_adjusted['inflation_rate'] / 100)
commodities_inflation_adjusted['cotton_value_adjusted'] = commodities_inflation_adjusted['cotton_value'] / (1 + commodities_inflation_adjusted['inflation_rate'] / 100)
commodities_inflation_adjusted['sugar_value_adjusted'] = commodities_inflation_adjusted['sugar_value'] / (1 + commodities_inflation_adjusted['inflation_rate'] / 100)
commodities_inflation_adjusted['wheat_value_adjusted'] = commodities_inflation_adjusted['wheat_value'] / (1 + commodities_inflation_adjusted['inflation_rate'] / 100)

# Display the updated dataframe with inflation-adjusted columns
commodities_inflation_adjusted.head(5)
````

Since inflation is compounded yearly, inflation is factored and commodity prices are compared over time relativfe to base year, 2000. The adjustment should account for cumulative inflation across mulitple years. This method is often reffered to as coverting prices to constant dollars (adjusting for inflation relative to base year). This provides a more accurate comparison of prices over time without the distortions caused by inflation.

Adjusted Value = (commodity value) / (1 + cumulative inflation)

````python
# Step 1: Create a cumulative inflation column starting from the base year (2000)
commodities_inflation_adjusted['cumulative_inflation'] = (1 + commodities_inflation_adjusted['inflation_rate'] / 100).cumprod()

# Step 2: Set the base year cumulative inflation to 1 (i.e., no adjustment needed for the base year)
commodities_inflation_adjusted.loc[2000, 'cumulative_inflation'] = 1

# Step 3: Adjust the commodity prices relative to the base year (2000) prices
commodities_inflation_adjusted['coffee_value_adjusted'] = commodities_inflation_adjusted['coffee_value'] / commodities_inflation_adjusted['cumulative_inflation']
commodities_inflation_adjusted['corn_value_adjusted'] = commodities_inflation_adjusted['corn_value'] / commodities_inflation_adjusted['cumulative_inflation']
commodities_inflation_adjusted['cotton_value_adjusted'] = commodities_inflation_adjusted['cotton_value'] / commodities_inflation_adjusted['cumulative_inflation']
commodities_inflation_adjusted['sugar_value_adjusted'] = commodities_inflation_adjusted['sugar_value'] / commodities_inflation_adjusted['cumulative_inflation']
commodities_inflation_adjusted['wheat_value_adjusted'] = commodities_inflation_adjusted['wheat_value'] / commodities_inflation_adjusted['cumulative_inflation']

# Display the updated dataframe with inflation-adjusted values
commodities_inflation_adjusted.head(5)
````

````python
plt.figure(figsize=(12, 6))
plt.plot(commodities_inflation_adjusted.index, commodities_inflation_adjusted['coffee_value_adjusted'], label='Coffee adjusted', color='brown')
plt.plot(commodities_inflation_adjusted.index, commodities_inflation_adjusted['corn_value_adjusted'], label='Corn adjusted', color='yellow')
plt.plot(commodities_inflation_adjusted.index, commodities_inflation_adjusted['cotton_value_adjusted'], label='Cotton adjusted', color='blue')
plt.plot(commodities_inflation_adjusted.index, commodities_inflation_adjusted['sugar_value_adjusted'], label='Sugar adjusted', color='orange')
plt.plot(commodities_inflation_adjusted.index, commodities_inflation_adjusted['wheat_value_adjusted'], label='Wheat adjusted', color='green')

plt.title('Commodity Adjusted Value, base year 2000')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
````

## Identifying Commodities Outliers

### Outlier Months for Coffee Prices

````python
# Calculate the IQR for coffee prices
Q1 = filtered_commodities_df['coffee_value'].quantile(0.25)
Q3 = filtered_commodities_df['coffee_value'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = filtered_commodities_df[(filtered_commodities_df['coffee_value'] < lower_bound) | (filtered_commodities_df['coffee_value'] > upper_bound)]

# Display the outliers
print("Outliers in Coffee Prices:")
print(outliers)
````
